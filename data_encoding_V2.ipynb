```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
```

```python
dir = r"D:\Fernando\Escritorio\CRISP-DM\produccion_textil_CRISP-DM\data\tidy_data.csv"
data = pd.read_csv(dir)
data.dtypes
```

```python
data["DATES"] = pd.to_datetime(data["DATES"], format="%d/%m/%Y")
data.dtypes
```

```python
data_scaled = data.copy(deep=True)
for column in range(len(data_scaled.columns)):
    if column > 2:
        data_scaled.iloc[:, column] = preprocessing.MinMaxScaler().fit_transform( #StandardScaler
            data_scaled.iloc[:, column].values.reshape(-1,1))
data_scaled.head()
```

```python
import pandas
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
data_encoded = encoder.fit_transform(data.iloc[:, 1:3])
feature_names = encoder.get_feature_names_out()
data_encoded = pd.DataFrame(data_encoded, columns=feature_names)
data_encoded
```

```python
dataset = pd.concat([data.iloc[:, 0], data_encoded, data_scaled.iloc[:, 3:12]], axis=1)
dataset
```

Modelo de machine learning

```python
X_multiple = dataset.iloc[:, 1:23]
Y_multiple = dataset.iloc[:, 23]
```

```python
from sklearn.model_selection import train_test_split
#Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X_multiple, Y_multiple, test_size=0.05)
```

```python
#Definimos el algoritmo a utilizar
from sklearn import linear_model
lr_multiple = linear_model.LinearRegression()
```

```python
#Entrenamos el modelo
lr_multiple.fit(X_train, y_train)
```

```python
#Realizamos una predicción
Y_pred_multiple = lr_multiple.predict(X_test)
```

```python
print(lr_multiple.score(X_train, y_train))
```
